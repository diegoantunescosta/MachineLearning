```python
import pandas as pd

base = pd.read_csv("/content/iris.data")
base
```

```python
previsores = base.iloc[:,0:4].values
classe = base.iloc[:, 4].values
```

```python
from sklearn .preprocessing import LabelEncoder 

label_classe = LabelEncoder()
```

```python
classe = label_classe.fit_transform(classe)
classe
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
previsores
```

```python
#treinamento e teste
from sklearn .model_selection import train_test_split

previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size = 0.20, random_state = 0)
```

```python
#classificação
from sklearn.ensemble import RandomForestClassifier

classificador = RandomForestClassifier(n_estimators = 100)
classificador.fit(previsores_treinamento, classe_treinamento)
previsoes = classificador.predict(previsores_teste)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

acuracia = accuracy_score(classe_teste, previsoes)
matriz = confusion_matrix(classe_teste, previsoes)
precisao = precision_score(classe_teste, previsoes, average='macro')
recall = recall_score(classe_teste, previsoes, average='macro')
```
